## **Seaborn — Main Categories of Plots**

Seaborn is a Python data visualization library built on top of Matplotlib.  
It provides beautiful, easy-to-use plots for statistical data visualization.  

Seaborn plots can be grouped into **6 main categories**:

### **1. Matrix Plots** 1
Used to visualize data in a **matrix (tabular) form**.  

- **`sns.heatmap()`** → Heatmap with numeric values shown as colors  
- **`sns.clustermap()`** → Heatmap with hierarchical clustering  

### **2. Relational Plots** 2
Used to show **relationships between two or more variables**.  

- **`sns.scatterplot()`** → Scatter plot for two variables  
- **`sns.lineplot()`** → Line plot for trends over continuous data  

### **3. Distribution Plots** 3
Used to explore the **distribution of a single variable** or compare distributions.  

- **`sns.histplot()`** → Histogram for distribution  
- **`sns.kdeplot()`** → Kernel Density Estimation curve
- **`sns.rugplot()`** → Small ticks along axis

### **4. Categorical Plots** 4
Used to **compare categories** and **visualize distribution across categories**.  

- **`sns.barplot()`** → Mean/aggregate values for categories  
- **`sns.countplot()`** → Frequency of categories  
- **`sns.boxplot()`** → Distribution + outliers by category  
- **`sns.violinplot()`** → Distribution shape + density   
- **`sns.swarmplot()`** → Non-overlapping points for categories  

### **5. Regression Plots** 5
Used to show **relationships with a regression model fit**.  

- **`sns.regplot()`** → Scatter plot with regression line  
- **`sns.lmplot()`** → Regression plots with additional grouping features  

### **6. Multi-Plot Grids** 6
Used to create **complex, multi-faceted visualizations**.  

- **`sns.pairplot()`** → Pairwise plots between numerical variables  
- **`sns.jointplot()`** → Plot for 2D relationships + distribution   

## **Conclusion**
- **Relational & Regression** → Show relationships between variables  
- **Categorical & Distribution** → Explore distributions and comparisons  
- **Matrix & Multi-Plot Grids** → Summarize and compare data in structured ways  

Seaborn makes it easier to choose the right visualization by organizing functions into these categories.
