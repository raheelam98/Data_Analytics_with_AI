# Levels of Measurement in Data  

In statistics and data analysis, variables are classified into four levels of measurement: **Nominal, Ordinal, Interval, and Ratio**.  
Each level provides different ways of analyzing and interpreting the data.  

---

## 1. Nominal Data  
**Definition:**  
Represents categories that are **mutually exclusive** and have **no inherent order or ranking**.  

**Examples:**  
- Colors (red, blue, green)  
- Types of fruit (apple, banana, orange)  
- Hair color (blonde, brown, black)  
- Types of transportation (car, bus, train)  

**Analysis:**  
- Can be analyzed using **frequencies**, **percentages**, and the **mode**.  

**Key Feature:**  
- Categories are **distinct** and **cannot be ordered**.  

---

## 2. Ordinal Data  
**Definition:**  
Represents categories that have a **meaningful order or ranking**, but the **intervals between categories are not equal**.  

**Examples:**  
- Education level (High School → Bachelor’s → Master’s → Doctorate)  
- Satisfaction ratings (Very Dissatisfied → Very Satisfied)  
- Income levels (Low → Medium → High)  
- Likert scales (Strongly Agree → Strongly Disagree)  

**Analysis:**  
- Can be analyzed using **frequencies**, **percentages**, the **mode**, and the **median**.  

**Key Feature:**  
- Categories can be **ordered**, but differences between them are **not measurable**.  

---

## 3. Interval Data  
**Definition:**  
Represents numerical values where **the difference between values is meaningful**, but there is **no true zero point**.  

**Examples:**  
- Temperature in Celsius or Fahrenheit  
- Calendar years (2000, 2010, 2020)  
- IQ scores  

**Analysis:**  
- Supports **addition and subtraction** (differences are meaningful).  
- Mean and standard deviation can be calculated.  

**Key Feature:**  
- **No absolute zero**, meaning ratios are not meaningful (e.g., 20°C is not “twice as hot” as 10°C).  

---

## 4. Ratio Data  
**Definition:**  
Represents numerical values with a **true zero point**, allowing for the comparison of both differences and ratios.  

**Examples:**  
- Height and weight  
- Age  
- Income  
- Distance and speed  

**Analysis:**  
- Supports all arithmetic operations (**addition, subtraction, multiplication, division**).  
- Can calculate **mean, median, mode, standard deviation, and ratios**.  

**Key Feature:**  
- Presence of a **true zero** makes ratios meaningful (e.g., 20kg is twice as heavy as 10kg).  

---

## 🔑 Quick Recap  

- **Nominal:** Categories without order (e.g., colors, fruits).  
- **Ordinal:** Ordered categories, unequal spacing (e.g., satisfaction levels).  
- **Interval:** Numeric, meaningful differences, no true zero (e.g., temperature).  
- **Ratio:** Numeric, meaningful differences and ratios, true zero (e.g., weight, income).
